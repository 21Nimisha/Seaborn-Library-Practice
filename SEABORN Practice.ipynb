# Visualizing distribution of dataset

distplot()  - draws histogram 

kdeplot() - it is draws inside histogram kde = kernel density estimator

jointplot()

rugplot()

```python
x = randn(100)
```

```python
sns.distplot(x)  # this is the distribution plot of x.
```

```python
sns.distplot(x, kde = False)
# now can see by writing kde = false , line removed from the graph
```

```python
sns.distplot(x , rug = True)
```

```python
sns.distplot(x , rug = True, bins = 30)
```

```python
# only rug estimator

sns.distplot(x, hist =False , kde = True)
```

```python
sns.kdeplot(x, shade = True)

# in kdeplot shade parameter use for shading the normal distribution plot.
```

```python
sns.kdeplot(x, shade = True, bw = 1)

# bw function change the smoothness , shape of the kde plot.
```

```python
sns.kdeplot(x, shade = True, bw =1 , cut = 0)

# cut function increase or decrease the area of graph.
```

```python
sns.kdeplot(x, shade = True, bw =1 , cut = 10)
```
